# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Box plot for each feature
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(x='Quantity', y='Unit Price', data=df)
plt.subplot(2, 2, 2)
sns.boxplot(x='Quantity', y='Total Price', data=df)
plt.subplot(2, 2, 3)
sns.boxplot(x='Unit Price', y='Total Price', data=df)
plt.tight_layout()
plt.show()
```

```python
# Distribution of each feature
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='Quantity', kde=True)
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='Unit Price', kde=True)
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='Total Price', kde=True)
plt.tight_layout()
plt.show()
```
